```python
df1 = df[df["YEAR"] != "1901-2015"]

df1  = df1[df["Parameter"] == "Actual"]
df1 = df1.groupby("YEAR").agg({"ANNUAL" : np.nanmean})

plt.figure()

plt.plot(df1, "-o", color = "r", ms = 4,alpha = 0.75)

plt.xlabel("Year")
plt.ylabel("Annual Rainfall (in mm)")
plt.title("Annual Rainfall from year 1901 to 2015 in India")
a = [0,20,40,60,80,100,120]
year = [1900,1920,1940,1960,1980,2000,2020]
plt.xticks(a,year)
plt.grid()
```

```python
df2 = df[df["YEAR"] == "1901-2015"]
df2 = df2[df2["Parameter"] == "Mean"]
df2.drop(columns = ["YEAR", "Parameter", "ANNUAL", "JF", "MAM", "JJAS", "OND"], inplace = True)
subdivision = df2["SUBDIVISION"]
df2.drop(columns = ["SUBDIVISION"], inplace = True)

plt.figure()
colors = ["blue", "green", "red", "cyan", "magenta", "yellow", "black", "blue", "green", "red", "cyan", "magenta"]
months = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"]

y_offset = np.zeros(df2.shape[0])

for i in range(df2.shape[1]):
    plt.barh(subdivision, df2[months[i]], left=y_offset, color=colors[i], label = months[i])
    y_offset = y_offset + df2[months[i]]

plt.legend(fontsize=5)
plt.xlabel("Rainfall (in mm)")
plt.ylabel("Subdivisions")
plt.yticks(fontsize=4)
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Flatten,Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
#scaler = MinMaxScaler(feature_range=(0,1))
new_df = np.array(new_df["data"])
```

```python
train_size = int(len(new_df)*.8)
test_size = len(new_df) - train_size
train,test = new_df[0:train_size], new_df[train_size:len(new_df)]
print(train.shape, test.shape)
```

```python
def to_sequence(new_df, seq_size=1):
    x = []
    y = []
    
    for i in range(len(new_df)-seq_size-1):
        window = new_df[i:(i+seq_size)]
        x.append(window)
        y.append(new_df[i+seq_size])
        
    return np.array(x), np.array(y)
```

```python
seq_size = 12;

trainx, trainy = to_sequence(train, seq_size)
testx, testy = to_sequence(test, seq_size)

print(trainx.shape, trainy.shape)
print(testx.shape, testy.shape)
```

```python
trainx = np.reshape(trainx, (trainx.shape[0],1,trainx.shape[1]))
testx = np.reshape(testx, (testx.shape[0],1,testx.shape[1]))
#trainx
```

```python
model = Sequential()
model.add(LSTM(2000, input_shape=(None, seq_size)))
model.add(Dropout(0.2))
model.add(Dense(500))
model.add(Dropout(0.2))
model.add(Dense(500))
model.add(Dropout(0.2))
model.add(Dense(400))
model.add(Dropout(0.2))
model.add(Dense(1))

```

```python
model.compile(loss="mean_squared_error", optimizer="adam")
model.summary()
```

```python
model.fit(trainx, trainy, verbose=2, epochs=1000)
```
