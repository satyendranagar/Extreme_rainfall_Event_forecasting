```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sub-division_rainfall_act_dep_1901-2015.csv")
df.shape
```

```python
df.head(10)
```

```python
df1 = df[df["YEAR"] != "1901-2015"]

df1  = df1[df["Parameter"] == "Actual"]
df1 = df1.groupby("YEAR").agg({"ANNUAL" : np.nanmean})

plt.figure()

plt.plot(df1, "-o", color = "r", ms = 4,alpha = 0.75)

plt.xlabel("Year")
plt.ylabel("Annual Rainfall (in mm)")
plt.title("Annual Rainfall from year 1901 to 2015 in India")
a = [0,20,40,60,80,100,120]
year = [1900,1920,1940,1960,1980,2000,2020]
plt.xticks(a,year)
plt.grid()
```

```python
df2 = df[df["YEAR"] == "1901-2015"]
df2 = df2[df2["Parameter"] == "Mean"]
df2.drop(columns = ["YEAR", "Parameter", "ANNUAL", "JF", "MAM", "JJAS", "OND"], inplace = True)
subdivision = df2["SUBDIVISION"]
df2.drop(columns = ["SUBDIVISION"], inplace = True)

plt.figure()
colors = ["blue", "green", "red", "cyan", "magenta", "yellow", "black", "blue", "green", "red", "cyan", "magenta"]
months = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"]

y_offset = np.zeros(df2.shape[0])

for i in range(df2.shape[1]):
    plt.barh(subdivision, df2[months[i]], left=y_offset, color=colors[i], label = months[i])
    y_offset = y_offset + df2[months[i]]

plt.legend(fontsize=5)
plt.xlabel("Rainfall (in mm)")
plt.ylabel("Subdivisions")
plt.yticks(fontsize=4)
```

```python
df3 = df2[["JAN", "FEB"]]
df3["JAN-FEB"] = df2["JAN"]+df2["FEB"]
df3["MAR-MAY"] = df2["MAR"] +df2["APR"] + df2["MAY"]
df3["JUN-SEP"] = df2["JUN"] + df2["JUL"] +df2["AUG"]+df2["SEP"]
df3["OCT-DEC"] = df2["OCT"] + df2["NOV"] + df2["DEC"]
df3.drop(columns = ["JAN", "FEB"], inplace = True)

plt.figure()
colors = ["blue", "orange", "green", "red"]
months = ["JAN-FEB", "MAR-MAY", "JUN-SEP", "OCT-DEC"]

y_offset = np.zeros(df3.shape[0])

for i in range(df3.shape[1]):
    plt.bar(subdivision, df3[months[i]], bottom=y_offset, color=colors[i], label = months[i])
    y_offset = y_offset + df3[months[i]]

plt.title("Rainfall in Subdivision of India")
plt.xlabel("Subdivisions")
plt.ylabel("Rainfall (in mm)")
plt.legend(fontsize=5)
#plt.grid()
plt.xticks(fontsize=4, rotation=90)
```
